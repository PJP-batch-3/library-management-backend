```python
#issues (borrow_date return_date)
for i in range(202101,202151):
  date=datetime.date(randint(2019,2021), randint(1,3),randint(1,28))
  x = randint(16,25)
  Enddate = date + timedelta(days=x)
  d = str(date)  
  r = str(Enddate) 
  print("update issues set borrow_date = " + '"' + d + '"'  " where id = " + str(i)+";" )
  print("update issues set return_date = " + '"' + r + '"'  " where id = " + str(i)+";" )
```
